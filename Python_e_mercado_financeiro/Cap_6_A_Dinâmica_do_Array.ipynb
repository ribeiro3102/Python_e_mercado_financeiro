### IMPORTAÇÃO DE BIBLIOTECAS ESPECÍFICAS PARA GRÁFICOS COM ARRAY

```python
import numpy as np
import matplotlib.pyplot as plt
from math import cos, exp

x = np.arange(0,20,0.1)
n = len(x)
y = np.zeros(n)

for i in range(n):
    y[i] = cos(x[i])*exp(-0.1*x[i])
   
plt.plot(x,y,'-k')
plt.grid(True)
plt.xlabel('x',fontsize=14)
plt.ylabel("y=f(x)",fontsize=14)
plt.title("f(x) = exp(-0.1x)*cos(x)")

soma = y.sum()
media = y.mean()
desvio = y.std()
minimo = y.min()
maximo = y.max()
ptomin = y.argmin()
ptomax = y.argmax()

print("=======ESTATÍSTICAS DO VETOR Y=======")
print("Soma: ",soma)
print("Média: ",media)
print("Desvio Padrão: ",desvio)
print("Minimo: ", minimo)
print("Máximo: ",maximo)
print("Índice do menor valor de y: ",ptomin)
print("Índice do maior valor de y: ",ptomax)
```

### Exemplo 6.2

```python
# Tendência Linear (o exemplo do livro é 01/01/2019 a 30/11/2019, 
# aqui foi usado o período total do ano, 01/01/2019 a 31/12/2019)
import xlrd as excel
import statistics as estat
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as m

wb = excel.open_workbook("dados/NTNB_2045_2019.xls")
plan = wb.sheet_by_name("Planilha1")

lin = plan.nrows
col = plan.ncols

op = np.zeros(lin)

for i in range(lin):
    op[i] = plan.cell(i,0).value

x = np.linspace(1,lin,lin)

coef = np.polyfit(x,op,1)

tendencia = coef[1] + coef[0]*x

print("=======Coeficiente de Tendência Linear y = b + ax =============")
print("a = ", coef[0], " , b = ",coef[1])

plt.plot(x, op, '-k',x, tendencia,'--k')
plt.xlabel("Dias")
plt.title("Títulos NTNB-principal 2045 (Selic + IPCA)")

filtro = m.detrend_linear(op)
plt.figure()
plt.plot(x, filtro,'-k')
plt.xlabel("Dias")
plt.title("Dados com remoção da linha de têndência linear")
```

### 6.7 PRODUTO, MATRIZ INVERSA E TRANSPOSTA DE UM ARRAY

```python
# Produto
import numpy as np

mat1 = np.array([[2,5],[3,-2]])
mat2 = np.array([[4,1],[-3,-7]])

prod = np.matmul(mat1, mat2)

print(prod)
```

```python
# Inversa
import numpy as np
from numpy.linalg import inv

A = np.array([[2,5],[3,-2]])

Inversa_A = inv(A)
Identidade = np.matmul(A, Inversa_A)
                       
print("Matriz A")
print(A)
print("")
print("Matriz inversa de A")
print(Inversa_A)
print("")
print("Matriz Identidade")
print(Identidade)
```

```python
# Transposta
import numpy as np

A = np.array([[2,5],[3,-2]])

Transposta_A = np.transpose(A)

print("Matriz A")
print(A)
print("")
print("Matriz inversa de A")
print(Transposta_A)
```

### 6.8 RESOLUÇÃO DE SISTEMA LINEAR COM ARRAY BIDIMENSIONAL

```python
# Transposta
import numpy as np
from numpy.linalg import inv,solve, det

A = np.array([[2,5],[3,-2]])
B = np.array([-1,5])

Inversa_A = inv(A)
X = np.matmul(Inversa_A,B)

print("Solução: ", X)
print("")
print("Solução usando a função solve ")
X2 = solve(A,B)
print(X2)
print("")
print("Descobrindo a determinante")
determinante = det(A)
print("Determinante: ", determinante)
```
