### 3.7 AJUSTE DE DISTRIBUIÇÕES DE PROBABILIDADE NORMAL AOS DADOS - PDF

```python
aleat = []

for i in range(1000):
    x = rd.gauss(0,1)
    aleat.append(x)
    
plt.subplot(211)
plt.plot(aleat, color="black")
plt.xlabel("Eixo Horizontal")
plt.ylabel("Números aleatórios entre -1 e 1")


plt.subplot(212)
plt.hist(aleat,bins=20, color="gray", density = True)
plt.xlabel("Classes")
plt.ylabel("Frequência")

xmin, xmax = plt.xlim()
media = st.mean(aleat)
desvio = st.pstdev(aleat)
eixo_x = np.linspace(xmin, xmax, 100)
eixo_y = norm.pdf(eixo_x, media, desvio)
plt.plot(eixo_x, eixo_y, color = "black")

print("Media: ", media)
print("SD: ", desvio)
```

### 3.9 ANÁLISE DE REGRESSÃO LINEAR NO MERCADO FINANCEIRO

```python
#===========TRANSFERE DADOS DA PLANILHA PARA O PRIMEIRO ATIVO=================
wb1 = excel.open_workbook("dados/IBOV7.xls")
plan1 = wb1.sheet_by_name("IBOV7")
lin1 = plan1.nrows
col1 = plan1.ncols
dados1 = []

#===========TRANSFERE DADOS DA PLANILHA PARA O SEGUNDO ATIVO=================
wb2 = excel.open_workbook("dados/PETR4_2.xls")
plan2 = wb2.sheet_by_name("PETR4")
lin2 = plan2.nrows
col2 = plan2.ncols
dados2 = []

for i in range(col1):
    coluna1 = plan1.col_values(i)
    dados1.append(coluna1)
    coluna2 = plan2.col_values(i)
    dados2.append(coluna2)
    
#=============TRANSFORMA LISTA EM VETOR
dados1 = np.array(dados1)
dados2 = np.array(dados2)

#============REGRESSÃO LINEAR================
beta, beta0, r_value, p_value, std_err = sci.linregress(dados1[1], dados2[1])
print("+++++++++++++++++++++++ REGRESSÃO LINEAR ++++++++++++++++++++++++++++++")
print("  beta      beta0      r_value      p_value      str_err")
yLin = beta*dados1[1] + beta0

print("%10.6f %10.6f %5.2f     %10.6f    %10.8f" % (beta, beta0, r_value, p_value, std_err))

plt.figure(figsize=(12, 6))
plt.plot(dados1[1], yLin, "-k", dados1[1], dados2[1],"ok")
plt.xlabel("IBOVESPA")
plt.ylabel("PETR4")
```

```python
# TESTE T PARA DUAS AMOSTRAS
# TRANSFERE DADOS DA PLANILHA PARA O PRIMEIRO ATIVO
wb1 = excel.open_workbook("dados/PETR4.xls")
plan1 = wb1.sheet_by_name("PETR4")
lin1 = plan1.nrows
col1 = plan1.ncols
dados1 = []

#===========TRANSFERE DADOS DA PLANILHA PARA O SEGUNDO ATIVO=================
wb2 = excel.open_workbook("dados/VALE3.xls")
plan2 = wb2.sheet_by_name("VALE3")
lin2 = plan2.nrows
col2 = plan2.ncols
dados2 = []

for i in range(col1):
    coluna1 = plan1.col_values(i)
    dados1.append(coluna1)
    coluna2 = plan2.col_values(i)
    dados2.append(coluna2)
    
#=============TRANSFORMA LISTA EM VETOR
dados1 = np.array(dados1)
dados2 = np.array(dados2)
retorno1 = (dados1[1,1:lin1]-dados1[1,0:lin1-1])/dados1[1,0:lin1-1]
retorno2 = (dados2[1,1:lin2]-dados2[1,0:lin2-1])/dados2[1,0:lin2-1]
# TESTE T
print(sci.ttest_ind(retorno1, retorno2))
plt.figure(figsize=(12, 6))
plt.subplot(311)
plt.plot(retorno1,"-k")
plt.title("RETORNO PETR4")

plt.subplot(312)
plt.plot(retorno2,"-k")
plt.title("RETORNO VALE3")

ax = plt.subplot(313)
ax.boxplot([retorno1,retorno2], showfliers=False)
ax.set_xticklabels(["PETR4","VALE3"])
```

### 3.11 COEFICIENTE DE CORRELAÇÃO

```python
# RELAÇÃO LINEAR E COEFICIENTE DE RELAÇÃO
wb1 = excel.open_workbook("dados/IBOV7.xls")
plan1 = wb1.sheet_by_name("IBOV7")
lin1 = plan1.nrows
col1 = plan1.ncols
dados1 = []

#===========TRANSFERE DADOS DA PLANILHA PARA O SEGUNDO ATIVO=================
wb2 = excel.open_workbook("dados/PETR4_2.xls")
plan2 = wb2.sheet_by_name("PETR4")
lin2 = plan2.nrows
col2 = plan2.ncols
dados2 = []

for i in range(col1):
    coluna1 = plan1.col_values(i)
    dados1.append(coluna1)
    coluna2 = plan2.col_values(i)
    dados2.append(coluna2)
    
#=============TRANSFORMA LISTA EM VETOR
dados1 = np.array(dados1)
dados2 = np.array(dados2)

# ============REGRESSÃO LINEAR
print("===========================REGRESSÃO LINEAR======================")

beta, beta0, r_value, p_value, std_err = sci.linregress(dados1[1],dados2[1])

print("   beta      beta0      r_value      p_value      std_err")

y_Lin = beta*dados1[1] + beta0

print("%10.6f  %10.6f  %5.2f  %10.6f  %10.8f" % (beta,beta0,r_value,p_value,std_err))

plt.figure(figsize=(12, 6))
plt.plot(dados1[1],y_Lin,"-k",dados1[1],dados2[1],"ok")
plt.xlabel("IBOVESPA",fontsize=14)
plt.ylabel("PETR4",fontsize=14)

#=================CORRELAÇÃO
cor,pval = sci.pearsonr(dados1[1],dados2[1])
plt.text(87000,18,"r = ", fontsize=14)
plt.text(90000,18,str(cor), fontsize=14)
print("===========================++++++++++++======================")
print("Correlação    p-value")
print(cor,pval)
```
