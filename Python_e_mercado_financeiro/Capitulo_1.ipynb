```python
# import xlrd as excel2
# import numpy as np
# import matplotlib.pyplot as plt
wb = excel2.open_workbook("dados/Petrobras.xls")
plan2 = wb.sheet_by_name("Planilha1")
x = plan2.col_values(0)

vetor = np.array(x)

print(vetor)
```

```python
# cálculo do retorno dos preços no Python é:
retorno = (vetor[1:20]-vetor[0:19])/vetor[0:19]

print(retorno)
```

```python
t = np.arange(0,19)
plt.title("Retorno")
plt.xlabel("Tempo")
plt.ylabel("Retorno")
plt.plot(t,retorno)
```

```python
import math as mt
media = st.mean(retorno)
print("Media: ", media)

desvio = st.pstdev(retorno)
print("Desvio: ", desvio)

otimista = media + 2*desvio/mt.sqrt(19)
print("Otimista: ", otimista*100)

pessimista = media - 2*desvio/mt.sqrt(19)
print("Pessimista: " , pessimista*100)
```
