### 12.4 ESTATÍSTICAS ENTRE PREÇOS E PALAVRAS-CHAVE

```python
import matplotlib.pyplot as plt
from pytrends.request import TrendReq
import matplotlib.dates as md
import pandas_datareader.data as pdr
import datetime as dt
import yfinance as yf
import pandas as pd
import scipy.stats as stats
import seaborn as sns

yf.pdr_override()

# conexão com o google
pytrends = TrendReq(hl='en-US', tz=360)

# palavras chave
kw_list=['petróleo']

# construção do gráfico
figura = plt.figure(figsize=(12,6))
ax1 = plt.subplot(111)
plt.title("GoogleTrends - Palavra chave")

#tipos de dados (anuais, mensais, diários, etc)
pytrends.build_payload(kw_list, cat=0,
                       timeframe="2018-07-01 2019-11-29",
                       geo='BR',gprop='')

teste=pytrends.interest_over_time()

# formatação da figura
ax1.plot(teste[kw_list[0]],'-k')
ax1.xaxis.set_major_formatter(md.DateFormatter('%d-%m-%Y'))
plt.setp(ax1.get_xticklabels(),rotation=45)
# Legenda
plt.legend((kw_list[0]),loc='upper left',shadow=True)

# identificação da palavra chave no gráfico
ax1.text(x=teste.index[-1],y=teste[kw_list[0]][-1],s=kw_list[0],fontsize=14,color='black',weight='bold')

# DataReader
inicio = dt.datetime(2018,7,1)
final = dt.datetime(2019,11,29)
df = pdr.get_data_yahoo("PETR4.SA",inicio,final)

ax2 = ax1.twinx()
ax2.plot(df.index,df['Close'],'--k')
ax2.set_ylabel('PETR4',fontsize=18,weight='bold')
ax2.text(x=teste.index[73],y=30,s='PETR4',fontsize=14,weight='bold')

# amostragem dos pontos da ação
med = df['Close'].resample('w').mean()
ax2.plot(med.index, med,'-k',linewidth=2)

# reta da regressão linear
plt.figure(figsize=(12,6))
dados1 = sns.regplot(x=teste['petróleo'].values,y=med,color='black')
dados1.set_xlabel('Palavra "petróleo"',fontsize=16)
dados1.set_ylabel('Preço PETR4 (R$)',fontsize=16)


# plotagem conjunta dos histogramas das séries
dados2 = sns.jointplot(x=teste['petróleo'].values,y=med,kind='scatter',color='black')
#dados2.set_axis_labels('Palavra "petróleo"', 'Preço PETR4 (R$)',fontsize=16)

# coeficiente de correlação de Pearson
#dados2.ax_joint.annotate(stats.pearsonr)

r, p = stats.pearsonr(teste['petróleo'].values, med)

dados2.ax_joint.annotate(f'$\\rho = {r:.5f}, p = {p:.5f}$',
                    xy=(0.1, 0.9), xycoords='axes fraction',
                    ha='left', va='center',
                    bbox={'boxstyle': 'round', 'fc': 'powderblue', 'ec': 'navy'})
dados2.ax_joint.scatter(teste['petróleo'].values, med)
dados2.set_axis_labels(xlabel='Palavra "petróleo"', ylabel='Preço PETR4 (R$)', size=15)
plt.tight_layout()
plt.show()
```
