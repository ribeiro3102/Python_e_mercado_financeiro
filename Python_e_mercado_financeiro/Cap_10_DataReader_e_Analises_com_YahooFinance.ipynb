```python
# Exemplo 10.7
import datetime as dt
import pandas_datareader.data as pdr
import pandas as pd
import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
yf.pdr_override()

inicio = dt.datetime(2019,1,1)
final = dt.datetime(2019,11,24)

df1 = pdr.get_data_yahoo('^bvsp',inicio,final)
df2 = pdr.get_data_yahoo('^dji',inicio,final)

# pareamento de datas para o mesmo dia
intersect = df2.index.intersection(df1.index)
bov = df1.loc[intersect]
dow = df2.loc[intersect]

correlacao = bov.corrwith(dow,axis=0)
dados= sns.regplot(x=bov['Close'], y=dow['Close'],color='black')

dados.set_xlabel('IBOVESPA',fontsize=16)
dados.set_ylabel('DOW JONES',fontsize=16)
dados.set_title('Regressão Linear Ibovespa x Dow Jones (Jan-Nov)/2019',fontsize=18)
print("   Correlação")
print(correlacao)
```

### 10.3 CÁLCULOS COM JANELAS MÓVEIS (*ROLLING METHOD*)

```python
# EXEMPLO 10.9
import matplotlib.pyplot as plt
import datetime as dt
import pandas_datareader.data as pdr
import yfinance as yf

plt.figure(figsize=(14,6))
yf.pdr_override()

inicio = dt.datetime(2015,1,1)
final = dt.datetime(2019,11,24)

df = pdr.get_data_yahoo('^bvsp',inicio,final)
df['Close'].plot(color='k',lw=2,alpha=0.6)
df['med_mov'] = df['Close'].rolling(window=150,min_periods=0).mean()
df['med_mov'].plot(color='k',lw=3,style='--')
plt.grid()
plt.title('Ibovespa 2015 / 2019', fontsize=18,weight='bold')
plt.legend(['Ibovespa','Média Móvel 150 dias'])
```

```python
# EXEMPLO 10.10
import matplotlib.pyplot as plt
import datetime as dt
import pandas_datareader.data as pdr
import yfinance as yf

plt.figure(figsize=(14,6))
yf.pdr_override()

inicio = dt.datetime(1995,1,1)
final = dt.datetime(2019,11,24)

df = pdr.get_data_yahoo('^bvsp',inicio,final)
df['retorno'] =df['Close'].pct_change()
df['retorno'].plot(color='k',lw=2, alpha=0.4)
df['std_mov'] = df['retorno'].rolling(window=10,min_periods=0).std()
df['std_mov'].plot(color='k',lw=3,style='-')

plt.grid()
plt.title('Retorno Financeiro Ibovespa (1995 - 2019)', fontsize=18,weight='bold')
plt.legend(['Ibovespa','Volatilidade(desv. padrão) 10 dias'])
```

### 10.4 VISUALIZAÇÃO COM *CANDLESTICK* DA BIBLIOTECA MPL_FINANCE

```python
# EXEMPLO 10.14
import matplotlib.pyplot as plt
import datetime as dt
import pandas_datareader.data as pdr
import yfinance as yf
import matplotlib.dates as mdt
from mpl_finance import candlestick_ohlc as mpl

plt.figure(figsize=(12,6))
yf.pdr_override()

inicio = dt.datetime(2019,1,1)
final = dt.datetime(2019,11,24)

df = pdr.get_data_yahoo('^bvsp',inicio,final)
#amostragem para os candles
df_ohlc = df['Close'].resample('7D').ohlc()
df_ohlc.reset_index(inplace=True)
df_ohlc['Date']=df_ohlc['Date'].map(mdt.date2num)

ax1 = plt.subplot(111)

ax1.xaxis_date()

mpl(ax1,df_ohlc.values, width=2, colorup='k',colordown='gray')
ax1.set_xlabel('DATA',fontsize=16)
ax1.set_ylabel('IBOVESPA',fontsize=16)
plt.grid()
```

```python
from yahoo_fin.stock_info import *
import numpy as np
import time as tm
import matplotlib.pyplot as plt
import datetime as dt
from mpl_finance import candlestick_ohlc as mpl
import matplotlib.dates as mdates
import pandas as pd
import pandas_datareader.data as pdr
import matplotlib.mlab as mlab
from scipy.stats import norm

lin = 10
x = np.zeros(lin)
y = np.zeros(lin)
xar=[]
yar=[]

plt.style.use('ggplot')
figura=plt.figure(figsize=(12,6))

ax1_din = plt.subplot(211)
ax2_din = plt.subplot(212)

for i in range(lin):
    x[i] = get_live_price('PETR4.SA')
    y[i] = get_live_price('USIM5.SA')
    print(i,round(x[i],2),round(y[i],2))
    xar.append(x[i])
    yar.append(y[i])
    ax1_din.plot(xar,'k')
    ax2_din.plot(yar,'k')
    plt.pause(0.5)
    tm.sleep(10)
    
ax1_din.set_title('PETR4')
ax2_din.set_title('USIM5')
```
