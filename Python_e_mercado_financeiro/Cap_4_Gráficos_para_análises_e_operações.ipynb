### GRÁFICO DE KDE *PLOT*

```python
aleat1=[]
for i in range(100):
    x= rd.gauss(0,10)
    aleat1.append(x)

aleat2=[]
for i in range(100):
    x= rd.gauss(0,10)
    aleat2.append(x)

plt.figure(figsize=(12,6))
plt.subplot(211)
plt.plot(aleat1,aleat2,'ok')

plt.subplot(212)
cmap = sns.cubehelix_palette(8, start=0,light=1,as_cmap=True)
#sns.kdeplot(x=aleat1,y=aleat2,cmap=cmap,fill=True)
sns.kdeplot(x=aleat1,y=aleat2,cmap=cmap,fill=True, alpha=0.7)
#shade=True foi deprecated (descontinuada), sendo indicado o uso de fill=True
```

### 4.5 GRÁFICO DINÂMICO

```python
plt.style.use('ggplot')

ax1=plt.subplot(211)
ax2=plt.subplot(212)
xar=[]
ret=[]
lin=10

x = np.zeros(lin)
retorno=np.zeros(lin)

for j in range(lin):
    x[j] = rd.gauss(0,1)
    if j>0:
        retorno[j]=(x[j]-x[j-1])/x[j-1]
    print('%d %5.2f'% (j,x[j]))
    xar.append(x[j])
    ret.append(retorno[j])
    ax1.plot(xar,'-ok')
    ax2.plot(ret,'-ok')
    plt.pause(0.5)
    tm.sleep(1)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d

ax = plt.axes(projection='3d')
z = np.linspace(0,15,1000)
x = np.sin(z)
y = np.cos(z)
ax.plot3D(x,y,z,'gray')

dadZ = 15*np.random.random(100)
dadX = np.sin(dadZ)+0.2*np.random.randn(100)
dadY = np.cos(dadZ)+0.2*np.random.randn(100)
ax.scatter3D(dadX,dadY,dadZ, c=dadZ, cmap='Greys');
ax.set_xlabel('eixo x')
ax.set_ylabel('eixo y')
ax.set_zlabel('eixo z')
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d

plt.figure(figsize=(12,8))
ax=plt.axes(projection='3d')

x = np.linspace(-20,20,50)
y = np.linspace(-20,20,50)

X, Y = np.meshgrid(x,y)
Z = np.sin(np.sqrt(X**2 + Y**2))
ax.contour3D(X,Y,Z,50, cmap='Greys')

ax.set_xlabel('eixo x')
ax.set_ylabel('eixo y')
ax.set_zlabel('eixo z')
```

```python
# FUNÇÃO plot_surface
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d

plt.figure(figsize=(12,8))
ax=plt.axes(projection='3d')

x = np.linspace(-2,2,50)
y = np.linspace(-2,2,50)

X, Y = np.meshgrid(x,y)
Z = np.sin(np.sqrt(X**2 + Y**2))
ax.plot_surface(X,Y,Z,cmap='Greys')

ax.set_xlabel('eixo x')
ax.set_ylabel('eixo y')
ax.set_zlabel('eixo z')
```
